```python
import pandas as pd

from omero_screen_analysis.countplot import PlotType, count_plot

df = pd.read_csv("../test_data.csv")
```

```python
conditions = ["ctr", "palb"]
```

```python
count_plot(
    df=df,
    norm_control="ctr",
    conditions=conditions,
    condition_col="condition",
    selector_col="cell_line",
    selector_val="RPE1wt",
    #plot_type = PlotType.NORMALISED, # this is the default!
    title="test rel counts 3 repeats",
    save=False,
)

count_plot(
    df=df,
    norm_control="ctr",
    conditions=conditions,
    condition_col="condition",
    selector_col="cell_line",
    selector_val="RPE1wt",
    plot_type=PlotType.ABSOLUTE,
    title="test abs counts 3 repeats",
    save=False,
)
```
