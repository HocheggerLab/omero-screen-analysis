```python
import pandas as pd

from omero_screen_analysis.featureplot import feature_plot
```

```python
df = pd.read_csv("../example_data_3x.csv")
```

```python
conditions = ["NT", "CCNA2", "CDK4"]
```

```python
feature_plot(
    df=df,
    feature="intensity_mean_p21_nucleus",
    conditions=conditions,
    condition_col="condition",  # default
    selector_col="cell_line",  # default
    selector_val="RPE-1_WT",
    title_str="p21",
    save=False,
    path=None,
)
```

```python
feature_plot(
    df=df,
    feature="area_cell",
    conditions=conditions,
    condition_col="condition",  # default
    selector_col="cell_line",  # default
    selector_val="RPE-1_WT",
    title="cell area",
    save=False,
    path=None,
)
```
