```python
import pandas as pd

from omero_screen_analysis.combplot import combplot

df = pd.read_csv("../example_data_3x.csv")

```

```python
conditions = ["siCTL DMSO", "siCTL 0.5uM", "siCTL 10uM"]
```

```python
combplot(
    df=df,
    conditions=conditions,
    feature_col="intensity_mean_p21_nucleus",
    feature_y_lim=6000,
    condition_col="condition",  # default
    selector_col="cell_line",  # default
    selector_val="HELA",
    title_str="test",
    cell_number=3000,
    save=False,
    path=None,
)
```

```python
from omero_screen_analysis.cellcycle_quant import cellcycle_plots

cellcycle_plots(
    df=df,
    conditions=conditions,
    condition_col="condition",  # default
    selector_col="cell_line",  # default
    selector_val="HELA",
    save=False,
    path=None,
)
```
